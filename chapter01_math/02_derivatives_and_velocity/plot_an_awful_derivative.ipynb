## Let's plot $t^3$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def car_position_eq(time):
    return time ** 3


time = np.arange(0, 4, 0.1)  # In seconds
car_positions = car_position_eq(time)  # In meters
```

```python
plt.plot(time, car_positions)
```

Now, let's show a marker for a given time and car position: $t_a=2.0$

```python
ta = 2.0
xa = car_position_eq(ta)

plt.plot(time, car_positions)
plt.plot(ta, xa, 'rs', fillstyle='none', markersize=12)
```

Now, let's calculate and plot the velocity for a $dt=1.0$. Remember that velocity is defined as
$$\vec{v} = \frac{x(t+dt) - x(t)}{dt}$$

```python
dt = 1.0
tb = ta + dt
xb = car_position_eq(tb)

plt.plot(time, car_positions)
plt.plot(ta, xa, 'rs', fillstyle='none', markersize=12)
plt.plot(tb, xb,'ro',fillstyle='none',markersize=10)
plt.xlim(0, max(time))
plt.ylim(0, max(car_positions))
```

Now, let's plot the line that joins the two points:
$$y-y_0 = m (x-x_0) \Rightarrow y = m (x-x_0) + y_0 $$

```python
# First, we need to plot the car position. Let's define a function for that
def plot_car_positions_and_points():
    plt.plot(time, car_positions)
    plt.plot(ta, xa, 'rs', fillstyle='none', markersize=12)
    plt.plot(tb, xb,'ro',fillstyle='none',markersize=10)
    plt.xlim(0, max(time))
    plt.ylim(0, max(car_positions))

# Now, let's call said function
plot_car_positions_and_points()

# And plot the new line
m = (xb - xa) / dt  # Did you realize? This is the velocity
v_line = m * (time - ta) + xa
plt.plot(time, v_line, 'g--')
plt.xlabel('Time (s)')
plt.ylabel('Car position (m)')
```
