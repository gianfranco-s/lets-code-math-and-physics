Import relevant libraries.  
`as` let's you use a shortcut for the imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

Create array from 0 to 10, with a step of 0.1

```python
time = np.arange(0, 10, 0.1)

# Alternatively, we could create an array from 0 to 10, with 10 elements
# time = np.linspace(0, 10, 10)
```

Calculate
$$y = \sin{t}$$

```python
y = np.sin(time)
```

Create a `plot` object.

```python
plt.plot(time, y)  # This action produces no output when run in a .py file

# Add some labels
plt.title('Sine wave')

plt.xlabel('Time (s)')
plt.ylabel('Amplitude')


# Show axii
plt.axhline(y=0, color='k',ls=':')
plt.axvline(x=0, color='k',ls=':')

# If running in a .py file, we'll need to run
# plt.show()
```
